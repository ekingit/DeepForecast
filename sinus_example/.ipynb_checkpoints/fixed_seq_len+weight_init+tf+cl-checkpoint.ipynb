# Pretrain

```python
class RNN_model_pretrain(nn.Module):
    def __init__(self, num_inputs, num_hiddens,num_out=1):
        super().__init__()
        self.num_inputs = num_inputs
        self.num_hiddens = num_hiddens
        self.rnn = nn.RNN(num_inputs, num_hiddens,bias=False, batch_first=True)
        self.linear = nn.Linear(num_hiddens, num_out)
    def forward(self, X, H):
        state, H = self.rnn(X,H)
        pred = self.linear(H[0]) #checked! H[0]=state[:,-1,:]
        return pred # size = (batch_size, 1)
```

```python
def rnn_train(model, X,Y, optimizer, loss, hidden_size, epoch, device = 'cpu'):
    total_loss = 0
    H = torch.zeros(1,X.shape[0],hidden_size)
    Y_hat = model(X,H)
    loss_out = loss(Y_hat,Y)
    loss_out.backward()
    optimizer.step()
    total_loss = torch.sqrt(loss_out.detach())
    if epoch %5000 == 0:
        print(total_loss) #MSE loss
```

```python
data = Data(0.1,200,0)
X,Y = data[15]
X = X.unsqueeze(-1)
```

```python
hidden_size = 2
mse_loss = nn.MSELoss()
rnn_model = RNN_model_pretrain(1,hidden_size)
optimizer = torch.optim.Adam(rnn_model.parameters(), lr=1e-4)
for epoch in range(50001):
    rnn_train(rnn_model, X,Y, optimizer, mse_loss, hidden_size, epoch)
```

```python
class RNN_model(nn.Module):
    def __init__(self, num_inputs, num_hiddens,num_out=1):
        super().__init__()
        self.num_inputs = num_inputs
        self.num_hiddens = num_hiddens
        self.rnn = nn.RNN(num_inputs, num_hiddens,bias=False,batch_first=True)
        self.linear = nn.Linear(num_hiddens, num_out)
        #custom weight initialization: 
        for i in range(2):
            self.rnn.all_weights[0][i].data = in_list[i]
        self.linear.weight.data = in_list[2]
        self.linear.bias.data = in_list[3]
    def forward(self, X, H):
        state, H_out = self.rnn(X,H)
        pred = self.linear(H_out[0])
        return pred
```
