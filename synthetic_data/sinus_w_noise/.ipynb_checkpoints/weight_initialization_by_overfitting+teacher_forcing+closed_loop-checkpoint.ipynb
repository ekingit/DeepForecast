```python
from data import Data
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

so far, we have learnt by experience that the same algorithm for the same data learn properly %80 of time. This is due to random weight initialization of the model. Usually, initialization depends on the specific problem at hand. 

Here, we will develop a general model which first learns right initialization with a learning algorithm (full teacher forcing), then employ our model (teacher forcing+closed loop).

```python
class RNN_model(nn.Module):  #@save
    def __init__(self, num_inputs, num_hiddens,num_out=1):
        super().__init__()
        self.num_inputs = num_inputs
        self.num_hiddens = num_hiddens
        self.rnn = nn.RNN(num_inputs, num_hiddens,bias=False)
        self.linear = nn.Linear(num_hiddens, num_out,bias=False)

    def forward(self, X, H, num, epoch, num_res=0):
        if num_res==0:
            num_res = len(X)
        
        pred_list = []
        if epoch < 100:
            out, H = self.rnn(X.reshape(len(X),1),H)
            pred_out = self.linear(out)
            pred_tens = pred_out.reshape(len(X))
        else:
            for i in range(num_res):
                if i < num:
                    out, H = self.rnn(X[i].reshape(1,1),H)
                    pred_out = self.linear(out)
                else:
                    out, H = self.rnn(pred_out, H)
                    pred_out = self.linear(out)
                pred_list.append(pred_out)
            pred_tens = torch.stack(pred_list).reshape(num_res)
        return pred_tens, H
```

```python
def train(model, X, optimizer, criterion, H, num, epoch):
    model.train()
    optimizer.zero_grad()
    y_hat,H = model(X,H,num, epoch)
    loss = criterion(X[1:],y_hat[:-1])
    loss.backward()
    optimizer.step()
    if epoch < 100:
        if epoch % 10 == 0:
            print(f"epoch: {epoch} loss {loss.detach()}")
    else:
        if epoch % 500 == 0:
            print(f"epoch: {epoch} loss {loss.detach()}")
```

```python
torch.manual_seed(4321)
data = Data(0.1,100,0.2)
X = data.x
H = torch.zeros(1,30)
mse_loss = nn.MSELoss()
model = RNN_model(1,30)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
for epoch in range(5000):
    H = torch.zeros(1,30)
    train(model, X, optimizer, mse_loss, H, 64, epoch)
```

```python
H = torch.zeros(1,30)
y, H = model(X, H, 50,1200,150)
plt.plot(X)
plt.plot(y.detach())
```
